```python
import numpy as np
import matplotlib.pyplot as plt
```

np.linspace范围 （-1,1），个数50

```python
x=np.linspace(-1,1,50)
y=2*x+1
```

plt.figure定义一个图像窗口

```python
plt.figure()
```

plt.plot画x,y曲线

```python
plt.plot(x,y)
```

plt.show显示图形

```python
plt.show()
```

```python
z=x**3+1
plt.figure(num=1,figsize=(8,5))#图像窗口编号为3，大小为（8,5）
plt.plot(x,z)
plt.plot(x,y,color='red',linestyle='--')#颜色为红色，类型为--
plt.show()
```

plt.xlim设置x坐标轴范围
plt.ylim设置y坐标轴范围
plt.xlabel设置x坐标轴名称
plt.ylabel设置y坐标轴名称

```python
plt.figure()
plt.xlim((-2,2))
plt.ylim((-2,2))
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,z)
plt.show()
```

plt.xticks设置x轴刻度
plt.yticks设置y轴刻度以及名称

```python
plt.xticks(np.linspace(-1,1,10))
plt.yticks(np.linspace(-1,1,4),['$good$','$very\ good$','$bad$','very bad'])
plt.show()
```

legend显示label信息

```python
plt.plot(x,y,label="linear line")
plt.plot(x,z,label="square line",color="red",linestyle="--",linewidth=1.0)
plt.legend(loc='upper right')
plt.show()
```
